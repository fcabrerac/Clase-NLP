```python
import re
import pandas as pd

from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')

from sklearn.feature_extraction.text import CountVectorizer
```

# Punto 1: Pre-Procesamiento

- `[11 pts]` Leer el archivo `dialogos.csv` usando `pandas`
- `[11 pts]` Crear una nueva columna con el texto en min√∫scula, sin caracteres especiales ni n√∫meros y sin palabras vacias

```python
#Archivo csv
#leer el archivo
df = pd.read_csv('C:/Users/Usuario/Downloads/Talleres_NLP/dialogos.csv',sep=";")
df
```

```python
df['locucion']=df['Locuci√≥n'].apply(lambda x: x.lower()) #convertir una cadena en minuscula
df['locucion'] = df['locucion'].apply(lambda x: re.sub(r"[\W\d_]+", " ",str(x)))#Sin caracteres especiales df y sin numeros
df['locucion'] = df['locucion'].apply(lambda x: ' '.join([palabra for palabra in x.split() if palabra not in (stopwords_sp)])) #Quitar Palabras vacias
df
```

# Punto 2: Representaci√≥n vectorial

- `[11 pts]` Crear una bolsa de palabras (BoW) del corpus usando la columna pre-procesada
- `[11 pts]` ¬øCu√°ntas palabras hay en el vocabulario? (Usando la funci√≥n de `sklearn`)

```python
count_vect = CountVectorizer()
bow_rep = count_vect.fit_transform(df['locucion'].values)
len(count_vect.vocabulary_)
```
